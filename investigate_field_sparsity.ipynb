```python
import pandas as pd
import numpy as np

import os

from smc_utils import drop_identifiers
```

```python
surveys = [0,1,2,3,4]
sems = ['Fall 2020', 'Spring 2021', 'Fall 2021', 'Spring 2022', 'Fall 2022']
```

```python
big_df = pd.read_excel(os.path.join("..", "Data", "SMC_DATA_clean.xlsx"))
big_df = drop_identifiers(big_df)
```

```python
big_df.head()
```

```python
cols = big_df.columns.tolist()[1:]

data = []

for c in cols:
    for v in surveys:
        row = [c, v]
        survey_df = big_df[big_df['SURVEY'] == v]
        for s in sems:
            sem_df = survey_df[survey_df['SEMESTER'] == s]
            record_count = len(sem_df)
            null_count = sem_df[c].isna().sum()
            row.append(record_count - null_count)
        data.append(row)
```

```python
pop_df = pd.DataFrame(data=data, columns=['Feature', 'Survey', 'FA20', 'SP21', 'FA21', 'SP22', 'FA22'])
```

```python
pop_df.to_excel(os.path.join("..", "Data", "data_population.xlsx"))
```
